```python
import pandas as pd
import numpy as np
import gzip
import os
```

<h1 style = "color: red"> Raw Data </h1>

```python
# Specify the folder containing the GZ files
folder_path = r'C:\Users\ferna\Documents\GitHub\anderworm\Project\ds_shenzou\raw'

# Initialize an empty list to store DataFrames
dfs = []

# Iterate through all files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.gz'):
        file_path = os.path.join(folder_path, filename)
        
        # Open and read the GZ file
        with gzip.open(file_path, 'rt') as gz_file:
            # Read the content of the GZ file into a DataFrame skipping the first row
            data = pd.read_csv(gz_file, delimiter='\t', low_memory=False, dtype=str, skiprows=[0])  # Skip the first row
            
            # Add a 'file_name' column with the current file's name
            data['file_name'] = filename
            
            # Append the DataFrame to the list
            dfs.append(data)

# Concatenate all DataFrames in the list
df_raw_data = pd.concat(dfs, ignore_index=True)

# Now, combined_data contains the data from all GZ files with a 'file_name' column
df_raw_data.head()
```

```python
import re
# Define a function to perform the transformation
def transform_seq_id(seq_id):
    if pd.notna(seq_id):  # Check if the value is not NaN
        # Use regular expressions to extract the relevant parts and format them
        match = re.match(r'^\w+_(\w+)_(\d+)_(\w+)_(\d+)$', seq_id)
        if match:
            gene_name = match.group(1).upper()
            gene_number = match.group(2)
            formatted_seq_id = f"{gene_name}-{gene_number} {match.group(3)} {match.group(4)}"
            return formatted_seq_id
    return seq_id  # Return the original value if it's NaN or doesn't match the pattern

# Apply the transformation to the 'SEQ_ID' column
df['SEQ_ID'] = df['SEQ_ID'].apply(transform_seq_id)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

import xgboost as xgb
df = pd.read_csv('base_teste.csv', low_memory=False)
df = df.rename(columns={'Factor Value: Genotype':'Genotype',
                 'Parameter Value: Absorbed Radiation Dose': 'Absorbed Radiation Dose',
                  'Factor Value: Altered Gravity':'Altered Gravity'})

df = df.dropna(subset=['PM'])
df['Altered Gravity G'] = np.where(df['Altered Gravity'] == 'uG', 10**-6, 1)

df['PM'] = df['PM'].astype(int)

df['Absorbed Radiation Dose mGy'] = df['Absorbed Radiation Dose'].str.split(' ').str[0].astype(float)

df['SEQ'] = df['SEQ_ID'].str.split(' ').str[0]

df['SEQ'] = df['SEQ'].str.split('-').str[0]

df = df[['PM', 'SEQ', 'POSITION','X','Y','Altered Gravity G','Absorbed Radiation Dose mGy','Genotype']]

df = pd.get_dummies(df,columns=['Genotype'], prefix='Genotype',drop_first=True, dtype=int)


# Calculate the frequency of each category using value_counts
category_counts = df['SEQ'].value_counts()/len(df['SEQ'])

# Create a new column 'Category_Frequency' by mapping values from 'Category' column
df['SEQ Category'] = df['SEQ'].map(category_counts)

df['SEQ Cluster'] = np.where(df['SEQ Category'] >= 0.004, df['SEQ'], 'Other')

df = df.drop(columns=['SEQ','SEQ Category'])

df = pd.get_dummies(df,columns=['SEQ Cluster'], prefix='SEQ',drop_first=True, dtype=int)


df_model = df.copy()

correlation_matrix = df_model.corr(method='kendall').round(2)
# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()

y = df_model['PM']
X = df_model.drop(columns=['PM'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

X_train, X_val, y_train, y_val  = train_test_split(X_train, y_train, test_size=0.25)

# Create an XGBoost regressor
xgb_regressor = xgb.XGBRegressor()

# Train the regressor on the training data
xgb_regressor.fit(X_train, y_train)

# Make predictions on the test data
y_pred = xgb_regressor.predict(X_test)

# Calculate the Mean Squared Error (MSE) of the model
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error (MSE): {mse:.2f}")
```
