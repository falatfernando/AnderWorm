```python
import pandas as pd
import numpy as np
import gzip
import os
```

<h1 style = "color: red"> Raw Data </h1>

```python
# Specify the folder containing the GZ files
folder_path = r'C:\Users\ferna\Documents\GitHub\anderworm\Project\ds_shenzou\raw'

# Initialize an empty list to store DataFrames
dfs = []

# Iterate through all files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.gz'):
        file_path = os.path.join(folder_path, filename)
        
        # Open and read the GZ file
        with gzip.open(file_path, 'rt') as gz_file:
            # Read the content of the GZ file into a DataFrame skipping the first row
            data = pd.read_csv(gz_file, delimiter='\t', low_memory=False, dtype=str, skiprows=[0])  # Skip the first row
            
            # Add a 'file_name' column with the current file's name
            data['file_name'] = filename
            
            # Append the DataFrame to the list
            dfs.append(data)

# Concatenate all DataFrames in the list
df_raw_data = pd.concat(dfs, ignore_index=True)

# Now, combined_data contains the data from all GZ files with a 'file_name' column
df_raw_data.head()
```

<h1 style = 'color: green'> Normalized Data <h1>

```python
# Specify the folder containing the GZ files
folder_path = r'C:\Users\ferna\Documents\GitHub\anderworm\Project\ds_shenzou\normalized'

# Initialize an empty list to store DataFrames
dfs = []

# Iterate through all files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.gz'):
        file_path = os.path.join(folder_path, filename)
        
        # Open and read the GZ file
        with gzip.open(file_path, 'rt') as gz_file:
            # Read the content of the GZ file into a DataFrame skipping the first row
            data = pd.read_csv(gz_file, delimiter='\t', low_memory=False, dtype=str, skiprows=[0])  # Skip the first row
            
            # Add a 'file_name' column with the current file's name
            data['file_name'] = filename
            
            # Append the DataFrame to the list
            dfs.append(data)

# Concatenate all DataFrames in the list
df_norm_data = pd.concat(dfs, ignore_index=True)

# Now, df_norm_data contains the data from all GZ files with a 'file_name' column
df_norm_data.head()
```
